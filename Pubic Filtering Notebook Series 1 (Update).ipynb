```python
import pandas as pd
import numpy as np
import string
```

```python
data = 'C:/Users/Alin/Desktop/pokemon.csv'  ##Input name of the File location and csv file here
df = pd.read_csv(data)
```

```python
df.head()####use to check data source loaded
```

```python
value_list =[
    'grass',
    'fire',
    'poison']  ##Input names of values you want to keep when filtering
```

```python
s = df['TYPE'].str.split(',',expand=True).stack().str.strip()   #####Input Column Name you want to use to filter values
```

```python
df_filter = df.loc[s[s.isin(value_list)].index.get_level_values(0)] ###Run to filter values(NOTE: ISIN may not work with your particular file so adjusments may need to be made )
###Additional aid - To handle complex strings loot toward using "str.contains"
```

```python
df_final = df_filter.dropna()  ###Remove nulls
```
